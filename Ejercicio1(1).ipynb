```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

# Parámetros de la distribución uniforme U(0, 1)
a = 0
b = 1

# Puntos para graficar la función de densidad y la CDF
x = np.linspace(-0.1, 1.1, 1000)
pdf = uniform.pdf(x, loc=a, scale=b-a)
cdf = uniform.cdf(x, loc=a, scale=b-a)

# Graficar la función de densidad
plt.plot(x, pdf, label='Función de Densidad')
plt.title('Distribución Uniforme U(0, 1) - Función de Densidad')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

# Graficar la distribución acumulada
plt.plot(x, cdf, label='Distribución Acumulada')
plt.title('Distribución Uniforme U(0, 1) - Distribución Acumulada')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.legend()
plt.show()

# Calcular media y varianza
media = uniform.mean(loc=a, scale=b-a)
varianza = uniform.var(loc=a, scale=b-a)

print(f'Media: {media}')
print(f'Varianza: {varianza}')
```

```python
import numpy as np
from scipy.stats import chi2

# Datos proporcionados
data = np.array([38, 33, 25, 5, 69, 35, 98, 52, 12, 79,
                 50, 46, 95, 42, 49, 11, 78, 34, 2, 43,
                 7, 50, 5, 91, 77, 18, 21, 4, 17, 62,
                 91, 36, 48, 88, 52, 76, 99, 73, 82, 90])

# Definir intervalos
num_intervals = 10
interval_limits = np.linspace(min(data), max(data), num_intervals+1)

# Contar frecuencia observada en cada intervalo
observed_freq, _ = np.histogram(data, bins=interval_limits)

# Calcular frecuencia esperada
expected_freq = len(data) / num_intervals

# Calcular chi^2
chi_squared = np.sum((observed_freq - expected_freq)**2 / expected_freq)

# Grados de libertad
df = num_intervals - 1

# Valor crítico para un nivel de significancia del 95%
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, df)

# Comparar con valor crítico
if chi_squared > critical_value:
    print(f'Rechazar la hipótesis nula. Chi^2 = {chi_squared} > Valor Crítico = {critical_value}')
else:
    print(f'No se puede rechazar la hipótesis nula. Chi^2 = {chi_squared} <= Valor Crítico = {critical_value}')
```

```python
import numpy as np
from scipy.stats import chi2

def chi_square_uniformity_test(data, num_intervals, alpha=0.05):
    # Definir intervalos
    interval_limits = np.linspace(min(data), max(data), num_intervals+1)

    # Contar frecuencia observada en cada intervalo
    observed_freq, _ = np.histogram(data, bins=interval_limits)

    # Calcular frecuencia esperada
    expected_freq = len(data) / num_intervals

    # Calcular chi^2
    chi_squared = np.sum((observed_freq - expected_freq)**2 / expected_freq)

    # Grados de libertad
    df = num_intervals - 1

    # Valor crítico para un nivel de significancia dado
    critical_value = chi2.ppf(1 - alpha, df)

    # Comparar con valor crítico
    if chi_squared > critical_value:
        result = f'Rechazar la hipótesis nula. Chi^2 = {chi_squared} > Valor Crítico = {critical_value}'
    else:
        result = f'No se puede rechazar la hipótesis nula. Chi^2 = {chi_squared} <= Valor Crítico = {critical_value}'

    return result

# Ejemplo de uso con los datos proporcionados anteriormente
data_example = np.array([38, 33, 25, 5, 69, 35, 98, 52, 12, 79,
                         50, 46, 95, 42, 49, 11, 78, 34, 2, 43,
                         7, 50, 5, 91, 77, 18, 21, 4, 17, 62,
                         91, 36, 48, 88, 52, 76, 99, 73, 82, 90])

result_example = chi_square_uniformity_test(data_example, num_intervals=10)
print(result_example)
```
