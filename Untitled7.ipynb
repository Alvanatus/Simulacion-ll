```python
import numpy as np
import matplotlib.pyplot as plt

def tiempo_llegada():
    return np.random.uniform(0, 30)

def esperar_maximo_10():
    llegada_persona1 = tiempo_llegada()
    llegada_persona2 = tiempo_llegada()
    espera = abs(llegada_persona1 - llegada_persona2)
    return espera <= 10

def experimento(iteraciones):
    espera_count = 0
    probabilidad_lista = []
    for i in range(1, iteraciones + 1):
        if esperar_maximo_10():
            espera_count += 1
        probabilidad = espera_count / i
        probabilidad_lista.append(probabilidad)
    return probabilidad_lista

# Realizamos el experimento con 2000 iteraciones
exp = experimento(2000)

# Graficamos la probabilidad en función del número de iteraciones
plt.plot(exp)
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de esperar como máximo 10 minutos')
plt.show()

# Calculamos la probabilidad final y su desviación estándar
probabilidad_final = exp[-1]
desviacion_estandar = np.std(exp)
print("Probabilidad final:", probabilidad_final)
print("Desviación estándar:", desviacion_estandar)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def tiempo_llegada():
    return np.random.uniform(0, 30)

def esperar_maximo_10():
    llegada_persona1 = tiempo_llegada()
    llegada_persona2 = tiempo_llegada()
    espera = abs(llegada_persona1 - llegada_persona2)
    return espera <= 10

def experimento(iteraciones):
    espera_count = 0
    probabilidad_lista = []
    for i in range(1, iteraciones + 1):
        if esperar_maximo_10():
            espera_count += 1
        probabilidad = espera_count / i
        probabilidad_lista.append(probabilidad)
    return probabilidad_lista

# Realizamos el experimento con 2000 iteraciones
exp = experimento(2000)

# Graficamos la probabilidad en función del número de iteraciones
plt.plot(exp)
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de esperar como máximo 10 minutos')
plt.show()

# Histograma
plt.hist(exp, bins=30, density=True, alpha=0.6, color='g')

# Calculamos el intervalo de confianza del 95%
media = np.mean(exp)
desviacion_estandar = np.std(exp)
n = len(exp)
intervalo = stats.norm.interval(0.95, loc=media, scale=desviacion_estandar/np.sqrt(n))

# Dibujamos el intervalo de confianza en la gráfica
plt.axvline(x=intervalo[0], color='r', linestyle='--', label='Intervalo de confianza 95%')
plt.axvline(x=intervalo[1], color='r', linestyle='--')

plt.xlabel('Probabilidad')
plt.ylabel('Frecuencia')
plt.title('Histograma de probabilidad de esperar como máximo 10 minutos')
plt.legend(loc='best')
plt.show()

print("Intervalo de confianza (95%):", intervalo)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def tiempo_llegada():
    return np.random.uniform(0, 30)

def esperar_maximo_10():
    llegada_persona1 = tiempo_llegada()
    llegada_persona2 = tiempo_llegada()
    espera = abs(llegada_persona1 - llegada_persona2)
    return espera <= 10

def experimento(iteraciones):
    espera_count = 0
    probabilidad_lista = []
    for i in range(1, iteraciones + 1):
        if esperar_maximo_10():
            espera_count += 1
        probabilidad = espera_count / i
        probabilidad_lista.append(probabilidad)
    return probabilidad_lista

# Realizamos el experimento con 2000 iteraciones
exp = experimento(2000)

# Graficamos la probabilidad en función del número de iteraciones
plt.plot(exp)
plt.xlabel('Iteraciones')
plt.ylabel('Probabilidad')
plt.title('Probabilidad de esperar como máximo 10 minutos')
plt.show()

# Histograma
plt.hist(exp, bins=np.linspace(0, 1, 30), density=True, alpha=0.6, color='g')

# Calculamos el intervalo de confianza del 95%
media = np.mean(exp)
desviacion_estandar = np.std(exp)
n = len(exp)
intervalo = stats.norm.interval(0.95, loc=media, scale=desviacion_estandar/np.sqrt(n))

# Dibujamos el intervalo de confianza en la gráfica
plt.axvline(x=intervalo[0], color='r', linestyle='--', label='Intervalo de confianza 95%')
plt.axvline(x=intervalo[1], color='r', linestyle='--')

plt.xlabel('Probabilidad')
plt.ylabel('Frecuencia')
plt.title('Histograma de probabilidad de esperar como máximo 10 minutos')
plt.legend(loc='best')
plt.xlim(0, 1)
plt.show()

print("Intervalo de confianza (95%):", intervalo)


# Calculamos la probabilidad final y su desviación estándar
probabilidad_final = exp[-1]
desviacion_estandar = np.std(exp)
print("Probabilidad final:", probabilidad_final)
print("Desviación estándar:", desviacion_estandar)
```
